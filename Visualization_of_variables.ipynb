```python

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

ages = np.random.randint(0, 100, size=100)

plt.figure(figsize=(10, 6))
plt.hist(ages, bins=10, color='blue', alpha=0.7, edgecolor='black')
plt.title('Age Distribution (Matplotlib)')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.show()

plt.figure(figsize=(10, 6))
sns.histplot(ages, bins=10, kde=True, color='blue', alpha=0.7)
plt.title('Age Distribution (Seaborn)')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.show()
```
